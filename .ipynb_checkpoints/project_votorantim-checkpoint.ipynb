## 1. Pré-Processamento

- Normalização: 

Normalização significa reduzir a escala de valores em que varia a série temporal. Isso é importante porque as redes neurais são sensíveis durante a fase de treinamento, podendo dar grandes ajustes a pesos específicos dependendo do valor de saída [buscar referência]. A ideia é alterar a escala dos dados para ter média aproximadamente 0 e um desvio-padrão de 1. Logo, em teoria, os dados devem variar entre -1 e 1. Na prática, a presença de alguns dados discrepantes (outliers) podem prejudicar o processo, sendo importante, então, um pré-tratamento dos dados.

```python
scaler = StandardScaler().fit(data)
normalzd = scaler.transform(data) # Dados normalizados
```

```python
plt.plot(normalzd)
plt.xlabel('Time')
plt.ylabel('Dow Jones Index')
plt.show()
```

A série normalizada se comporta de forma idêntica à série original, porém numa escala menor.

- Criação de Janelas:

Janela é o nome dado a cada linha da matriz, que indica um trecho da série temporal. A ideia é transformar um vetor unidimensional de tamanho N, para uma matriz bidimensional que tem nas colunas os valores das entradas da rede, em ordem de tempo, e sua respectiva saída. Resumindo, na última coluna está o valor "futuro" e nas colunas anteriores, os valores antecessores. Logo, o tamanho das janelas ilustradas abaixo é igual a 3 (d = 3). A matriz ficou de tamanho N-d+1, d.

<img src='sketch1.png' style="width: 500px;">

```python
d = 3
n = data.shape[0]-d+1

matrix = np.zeros((n, d))
for i in range(n):
    for j in range(d):
        matrix[i,j] = data[i+j]
print('Série Original (5 primeiros valores) \n', data[:5].reshape(1,-1),'\n')
print('Matriz gerada (3 primeiras janelas)\n', matrix[:3, :])
```

- Seleção de Lags e Particionamento:

Muito provavelmente, isso não será colocado no trabalho. Mas a ideia é basicamente calcular a autocorrelação da série temporal e verificar se trechos específicos da série possuem comportamento semelhante entre si. Caso sim, seleciona-se pontos específicos da série e acrescenta-se como entrada da rede neural.

## 2. Modelagem/Treinamento

- Definição de parâmetros:

Os parâmetros podem ser o número de neurônios na camada oculta, o tamanho da janela que será utilizada e a função de ativação da rede neural (sigmoide, tangente hiperbólica etc).

- Gridsearch:

Criar redes neurais iterativamente explorando cada um dos parâmetros citados acima. Logo, serão criadas diversas MLPs combinando os 3 parâmetros entre si. 

Exemplo: 

Número de neurônios na camada oculta: [2, 3, 4, 5, 6, 7, 8, 9, 10] <br>
Tamanho da janela: [3, 4, 5, 6 ... fica a critério do tamanho da base de dados] <br>
Funções de ativação: [Tangente hiperbólica, Sigmoide]

- Critérios de parada:

Os critérios de parada do treinamento da MLP são: erro global e número de épocas. Não precisa se preocupar tanto com isso, pois já vem implementado na biblioteca. Mas é bom saber quais são os valores pra especificar no TCC.

- Escolha de um Modelo:

O modelo da rede neural será um Multi Layer Perceptron (MLP) com apenas uma camada oculta. Lembrando que saída é o valor "futuro" que se deseja calcular e as entradas (features) são os valores antecessores. É importante que a ordem das features siga a ordem do tempo. Logo, se a saída é referente ao tempo N+1, as entradas devem ser ordenadas em [N-2, N-1, N]. O número de features será igual a d-1.

<img src='sketch2.png' style="width: 300px;">

- Conjunto de treinamento:

A série temporal deverá ser separada em 3 conjuntos distintos: treinamento, validação e teste. Para o treinamento, poderá ser utilizado 50 ou 60% inicial da série.

<img src='sketch3.png' style="width: 700px;">

## 3. Avaliação

- Definição de Métricas:

À medida que as redes neurais forem sendo treinadas, deverão ser armazenados algumas métricas de avaliação do modelo. No final do processo de construção dos modelos, as métricas serão comparadas e busca-se os parâmetros que obteve melhor desempenho. Algumas métricas são destacadas à seguir [buscar referência].

<img src='sketch4.png' style="width: 400px;">

- Conjunto de validação:

O conjunto de validação serve para se extrair as métricas de avaliação de cada modelo. A ideia é usar um conjunto diferente do treinamento a fim de evitar o problema de overfitting [buscar referência]. O tamanho do conjunto de validação indicado é entre 20 e 25% da série temporal.

### Etapas A e B

O processo de treinamento e avaliação estão dentro de um loop em que se treina diferentes redes neurais e são armazenados os respectivos valores das métricas.
